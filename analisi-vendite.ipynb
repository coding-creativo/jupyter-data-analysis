## OPERAZIONI COMUNI

## RAGGRUPPAMENTO

```python
#Quando si utilizza groupby in Pandas, il risultato restituito non è direttamente stampabile come una tabella di dati. 
#Si ottiene un oggetto DataFrameGroupBy che rappresenta un oggetto raggruppato. 
#Per ottenere risultati aggregati, come la media dei prezzi per ogni tipo di pizza, è necessario applicare una funzione di aggregazione, 
#come ad esempio mean(), sum(), count(), ecc., 
#infatti se eseguo questo codice ottengo un oggetto DataFrameGroupBy 
prezzo_medio_per_tipo = df.groupby('type')
prezzo_medio_per_tipo
```

```python
# Raggruppare per tipo di pizza e calcolare il prezzo medio delle pizze
prezzo_medio_per_tipo = df.groupby('type')['price'].mean()
prezzo_medio_per_tipo
```

```python
df.head()
```

```python
# Filtra le righe dove la dimensione è "M" e poi conta le occorrenze per ogni tipo
pizze_size_M = df[df['size'] == 'M'].groupby('type').size()
pizze_size_M
```

```python
# Filtra le righe dove la dimensione è "M" e poi conta le occorrenze per ogni tipo
pizze_size_M = df[df['size'] == 'M'].groupby('type').size().reset_index(name='count_dim_M')
pizze_size_M
```

```python
# Conta le pizze vendute per ogni dimensione
pizze_per_size = df.groupby('size').size().reset_index(name='count')
pizze_per_size
```

```python
# Raggruppa per type e size e conta le occorrenze
sales_by_type_and_size = df.groupby(['type', 'size']).size().reset_index(name='count')
print(sales_by_type_and_size)
```

## ORDINAMENTO

```python
# Conta il numero di occorrenze di ogni tipo di pizza e ordina in ordine decrescente
pizza_counts = df['PizzaName'].value_counts()
#cambiamo l'ordine
pizza_counts_sorted = pizza_counts.sort_values(ascending=True)
pizza_counts_sorted
```
