# Matplotlib
Matplotlib è una libreria di visualizzazione di dati per il linguaggio di programmazione Python. È molto potente e offre una vasta gamma di grafici e plot per la creazione di visualizzazioni di dati di alta qualità. È utilizzato ampiamente in ambiti come la scienza dei dati, l'analisi statistica, l'ingegneria e molti altri settori che richiedono la rappresentazione visuale dei dati.

Matplotlib può creare tanti tipi di grafici, tra cui:

- ***Grafici a linee***: per rappresentare relazioni tra dati continui.
- ***Scatter plot***: per visualizzare punti di dati senza connessioni lineari.
- ***Grafici a torta***: per esprimere proporzioni o percentuali.
- ***Grafici a barre***: per confrontare quantità tra diverse categorie.
- ***Istogrammi***: per mostrare la distribuzione dei dati.


## Grafico a Linee

Per creare un grafico a linee in Matplotlib, possiamo utilizzare la funzione ***plt.plot()***. 

Passi principali per creare un grafico a linee:

* Definiamo i dati: Prepariamo i dati da visualizzare come linee. Di soluto due array o liste, uno per l'asse X e uno per l'asse Y, che rappresentano le coordinate dei punti nel grafico.

* Chiamata a plt.plot(): Utilizziamo plt.plot() passando i dati sull'asse X e sull'asse Y come argomenti principali. Si possono anche specificare opzioni aggiuntive come il colore della linea, lo stile della linea, i marcatori e tanto altro.

* Personalizzazione del grafico: Aggiungiamo etichette agli assi, un titolo al grafico, una legenda se necessario e abilitiamo la griglia per migliorare la leggibilità del grafico.

* Mostriamo il grafico: Utilizziamo plt.show() per visualizzare il grafico.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## ESEMPIO 1

```python
# 1) GRAFICO A LINEE
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creazione del grafico a linee
plt.figure(figsize=(8, 4))  
''' Definizione delle dimensioni della figura - figure è un parametro opzionale della funzione plt.figure() che specifica le dimensioni 
    della figura in pollici. Nel caso specifico, (8, 4) indica che la larghezza della figura sarà di 8 pollici e l'altezza sarà di 4 pollici.
'''
plt.plot(x, y, marker='o', linestyle='-', color='y', label='Dati')  
''' Creiamo un grafico a linee che rappresenta i dati forniti dalle liste x e y 
    Specifichiamo:
        1) Il tipo di marcatore da utilizzare sui punti del grafico. 
            In questo caso, 'o' indica che si vogliono i cerchi come marcatori per ciascun punto.
            'o': Cerchio, 's': Quadrato, '^': Triangolo verso l'alto, 'v': Triangolo verso il basso, '>': Triangolo verso destra, 
            '<': Triangolo verso sinistra, 'x': X, '+': Più, 'D': Diamante, 'p': Pentagono, '*': Stella, 
        2) linestyle='-'. Specifica lo stile della linea utilizzata per connettere i punti. 
            '-' indica una linea continua; '--' Linea tratteggiata; ':' Linea a puntini
        3) color='b' nomi comuni (come 'b' per blu, 'r' per rosso, ecc.) o codici esadecimali ('#FFA500' per arancione).
        4) label='Dati': Questo è il nome che verrà utilizzato per identificare la serie di dati nel caso in cui si desideri 
            aggiungere una legenda al grafico.
'''
plt.xlabel('X')  # Etichetta asse X
plt.ylabel('Y')  # Etichetta asse Y
plt.title('Grafico a Linee')  # Titolo del grafico
plt.legend()  # Mostra la legenda
plt.grid(True)  # Mostra la griglia
plt.show()  # Mostra il grafico
```

#### STESSO ESEMPIO CON NUMPY

```python
#2) STESSO ESEMPIO CON NUMPY
import numpy as np
x = np.arange(0,5) #valori da 0 a 4
y = np.arange(0,10,2) #valori 0, 2, 4, 6, 8

# Creazione del grafico a linee
plt.figure(figsize=(8, 4))  # Definizione delle dimensioni della figura
plt.plot(x, y, marker='s', linestyle='-', linewidth=2, markersize=10,  color='y', markerfacecolor='yellow', markeredgecolor='black', markeredgewidth=2, alpha=0.7, label='Dati') 
''' linewidth (o lw): Imposta lo spessore della linea che connette i punti nel grafico.
    Possibili valori: Numeri positivi che rappresentano lo spessore della linea in punti (pixels).
    Esempio: linewidth=2 imposta una linea con uno spessore di 2 punti.

    markersize (o ms): Determina la dimensione dei marcatori che rappresentano i punti nel grafico.
    Possibili valori: Numeri positivi che indicano il diametro del marcatore in punti.
    Esempio: markersize=10 imposta marcatori con un diametro di 10 punti.

    alpha: Regola la trasparenza dei marcatori e della linea nel grafico.
    Possibili valori: Numeri compresi tra 0 (completamente trasparente) e 1 (completamente opaco).
    Esempio: alpha=0.7 imposta una trasparenza del 70% per sia i marcatori che la linea.
    
    markerfacecolor (o mfc): Specifica il colore di riempimento dei marcatori nel grafico.
    Possibili valori: Stringhe che rappresentano nomi di colori (come 'yellow', 'blue', ecc.) o codici esadecimali.
    Esempio: markerfacecolor='yellow' riempie i marcatori con il colore giallo.

    markeredgecolor (o mec): Imposta il colore del bordo dei marcatori nel grafico.
    Possibili valori: Stringhe che rappresentano nomi di colori (come 'black', 'red', ecc.) o codici esadecimali.
    Esempio: markeredgecolor='black' imposta il bordo dei marcatori di colore nero.
    
    markeredgewidth (o mew): Controlla lo spessore del bordo dei marcatori nel grafico.
    Possibili valori: Numeri positivi che rappresentano lo spessore del bordo in punti.
    Esempio: markeredgewidth=2 imposta lo spessore del bordo dei marcatori a 2 punti.
'''
plt.xlabel('X', fontsize=12, fontstyle='italic')  # Etichetta asse X
plt.ylabel('Y', fontsize=12, fontweight='bold')  # Etichetta asse Y
plt.title('Grafico a Linee',fontsize=22)  # Titolo del grafico
plt.legend()  # Mostra la legenda
plt.grid(True)  # Mostra la griglia
plt.show()  # Mostra il grafico
```

## INFORMAZIONI AGGIUNTIVE

### LIMITI DEGLI ASSI

```python
# Limiti degli assi (xlim e ylim). Si possono definire i limiti degli assi X e Y per concentrare il grafico su un'area specifica. 
# È utile quando si desidera esaminare dettagli specifici dei dati.
plt.xlim(0, 6)  # Limiti dell'asse X da 0 a 6
plt.ylim(0, 12)  # Limiti dell'asse Y da 0 a 12
plt.show()  # Mostra il grafico
```

### PERSONALIZZARE LINEE GUIDA

```python
plt.grid(True, linestyle=':', linewidth=2.5, color='gray', alpha=0.5)
plt.show() 
```

### LEGENDA 

```python
plt.plot(x, y, marker='s', linestyle='-', color='y', label='Dati')  
plt.legend(fontsize='medium', loc='right')
'''
    fontsize --> specifica la dimensione del carattere in punti
             --> si usa per Etichette degli assi(xlabel, ylabel), Titoli(title), Legende(legend), Annotazioni o testi(text), 
                Tick labels (etichette degli intervalli, xticks, yticks)
    loc --> specifica la posizione di elementi come legende, titoli, e annotazioni all'interno di un grafico. 
        --> si usa  per legende, titoli degli assi, annotazioni o testi, colorbars (legenda della scala dei colori)
        
        Posizioni principali:

        'best': Trova automaticamente la migliore posizione per l'elemento.
        'upper right' o 'right': In alto a destra del grafico.
        'upper left' o 'left': In alto a sinistra del grafico.
        'lower right': In basso a destra del grafico.
        'lower left': In basso a sinistra del grafico.
        'center right': Alla destra del centro del grafico.
        'center left': Alla sinistra del centro del grafico.
        'center': Al centro del grafico.
        'upper center': In alto al centro del grafico.
        'lower center': In basso al centro del grafico.

        Posizioni specifiche:

        (x, y): Coordinate specifiche (in frazioni del grafico o in coordinate assolute) per posizionare l'elemento.
        'none': Non mostrare l'elemento.
        Numeri interi da 1 a 10 per posizioni relative in layout di griglia predefiniti.
        
        Annotazioni specifiche:

        'outside': Mostra la legenda al di fuori dell'area del grafico.
        'centered': Centra la legenda rispetto al punto specificato.
'''
plt.show()  
```

## DUE RETTE NELLO STESSO GRAFICO E USO DI plot_date()

```python
# 5) La funzione plot_date è utilizzata per tracciare dati in cui l'asse X rappresenta date o tempi.
# Accetta due argomenti principali: un array di date (asse X) e un array di valori (asse Y). Gli array devono essere della stessa lunghezza.
# plot_date interpreta automaticamente l'asse X come date, consentendo di visualizzare i dati temporali in modo appropriato. 
# chiaramente se si volessero le date sull'asse Y basta invertire i valori
# Matplotlib gestirà automaticamente la formattazione dell'asse X per mostrare le date in modo chiaro e leggibile.
# Come con altre funzioni di tracciamento in Matplotlib, è possibile personalizzare l'aspetto del grafico utilizzando parametri 
# come marker, linestyle, color, markersize, label, ecc.
```

```python
from datetime import datetime

# Dati di esempio
date = [datetime(2024, 6, 1), datetime(2024, 6, 2), datetime(2024, 6, 3), datetime(2024, 6, 4), datetime(2024, 6, 5)]
azioni_A = np.array([100, 105, 110, 108, 112])
azioni_B = np.array([95, 97, 98, 96, 100])

# Tracciamento delle due serie di dati (rette) utilizzando plt.plot_date()
plt.plot_date(date, azioni_A, 'o-', color='b', label='Azione A')
plt.plot_date(date, azioni_B, 's--', color='r', label='Azione B')

# Etichette degli assi, titolo e legenda
plt.xlabel('Data', labelpad=15)
plt.ylabel('Valore Azioni', labelpad=15)
plt.title('Andamento Azioni A e B')
plt.legend()

# Ruota le etichette sull'asse X per migliorare la leggibilità
plt.xticks(rotation=90)

# questa istruzione formatta la data
#plt.matplotlib.dates.DateFormatter('%d-%m-%Y')
# ma occorre formattare anche le etichette delle date
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d-%m-%Y'))  # Formato 'anno-mese-giorno'
''' 
plt.gca(): è un metodo che restituisce l'oggetto "Current Axes" (gli assi attualmente attivi) 
nel sistema di coordinate corrente di Matplotlib. "Current Axes" si riferisce agli assi su cui stai lavorando.

xaxis: Una volta ottenuto l'oggetto "Current Axes", xaxis rappresenta l'asse X di questi assi.

set_major_formatter(): Questo metodo permette di impostare un formattatore per le etichette sull'asse principale (major ticks) dell'asse X.

plt.matplotlib.dates.DateFormatter('%d-%m-%Y'): DateFormatter è una classe di Matplotlib che formatta le etichette delle date sull'asse X secondo il formato specificato. Nel caso specifico, '%d-%m-%Y' indica il formato delle date come giorno-mese-anno (es. 01-01-2023).
'''

plt.xticks(date)

plt.grid(True)
plt.tight_layout()  # Ottimizza il layout per evitare sovrapposizioni
plt.show()
```

```python
# 6) plot da un dataframe
import pandas as pd

np.random.seed(13)  #per riprodurre gli stessi valori

dates = pd.date_range(start='2023-01-01', periods=20)
dates
```

```python
cities = ['Roma', 'Milano', 'Torino']

df = pd.DataFrame({
    'Date': np.repeat(dates, len(cities)),
    'City': np.tile(cities, len(dates)), # creare una nuova array replicando un array di input
    'Temperature': np.random.randint(low=0, high=30, size=len(dates) * len(cities))
})

# Visualizziamo le prime righe del dataframe
df.head()
```

```python
# Creazione del grafico a linee per l'andamento delle temperature nelle città
plt.figure(figsize=(12, 8))

# Ciclo per creare una linea per ogni città
for city in cities:
    city_data = df[df['City'] == city]
    plt.plot(city_data['Date'], city_data['Temperature'], marker='o', linestyle='-', label=city)

# Personalizzazione del grafico
plt.title('Andamento delle temperature nelle città')
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.grid(True)
plt.legend()

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
