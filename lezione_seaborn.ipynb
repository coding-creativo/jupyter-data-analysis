### Grafico a dispersione (scatterplot)

Casi d'utilizzo:
* Relazione tra altezza e peso di un gruppo di persone. * Relazione tra spese pubblicitarie e vendite: Utilizzato in marketing per vedere se l'aumento della spesa pubblicitaria si traduce in maggiori vendite.* 
Relazione tra temperatura e consumo energeticoe.


```python

# Generazione di dati casuali per altezza e peso
np.random.seed(0)
altezza = np.random.normal(loc=170, scale=10, size=100)
peso = altezza * 0.6 + np.random.normal(loc=0, scale=7, size=100)

# Creazione del grafico a dispersione
plt.figure(figsize=(10, 6))

sns.scatterplot(x=altezza, y=peso, color='skyblue',  alpha=0.8, hue=altezza, size=peso, palette='viridis')

plt.title('Relazione tra Altezza e Peso')
plt.xlabel('Altezza (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Dati di esempio: vendite mensili per quattro prodotti (A, B, C, D)
mesi = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile', 'Maggio']
vendite_prodotto_A = [15000, 18000, 21000, 19000, 20000]
vendite_prodotto_B = [12000, 15000, 17000, 16000, 18000]
vendite_prodotto_C = [10000, 11000, 13000, 12000, 14000]
vendite_prodotto_D = [9000, 9500, 11000, 10500, 12000]

# Creazione del dataframe per Seaborn
import pandas as pd
df = pd.DataFrame({
    'Mese': mesi,
    'Prodotto A': vendite_prodotto_A,
    'Prodotto B': vendite_prodotto_B,
    'Prodotto C': vendite_prodotto_C,
    'Prodotto D': vendite_prodotto_D
})

# Impostazioni grafiche di Seaborn
plt.figure(figsize=(10, 6))  # Imposta le dimensioni della figura (larghezza, altezza)

# Tracciamento del grafico a barre con Seaborn
sns.barplot(x='Mese', y='value', hue='variable', data=pd.melt(df, ['Mese']), palette='muted')

# Etichette degli assi e titolo
plt.xlabel('Mesi')
plt.ylabel('Vendite')
plt.title('Vendite Mensili per Prodotto')

# Mostra la legenda
plt.legend(title='Prodotto', loc='upper left')

# Mostra il grafico
plt.grid(True, axis='y')  # Abilita le linee di griglia sull'asse Y
plt.tight_layout()  # Ottimizza il layout per evitare sovrapposizioni
plt.show()
```


### kde
kde, acronimo di Kernel Density Estimation, Ã¨ una tecnica statistica utilizzata per stimare la densitÃ  di probabilitÃ  di una variabile casuale. Nei grafici statistici, come gli istogrammi o i violin plot, il KDE Ã¨ spesso utilizzato per fornire una rappresentazione liscia e continua della distribuzione dei dati.

kde cerca di modellare la distribuzione sottostante dei dati, fornendo una rappresentazione piÃ¹ dettagliata rispetto agli istogrammimi

```python
# Generazione di dati casuali per le altezze (simulazione di una distribuzione normale)
np.random.seed(0)
altezze = np.random.normal(loc=170, scale=10, size=1000)  # Media 170 cm, deviazione standard 10 cm, 1000 campioni
#print(altezze)
# Creazione del grafico con histplot e KDE sovrapposto
plt.figure(figsize=(10, 6))

sns.histplot(altezze, kde=True, color='skyblue', edgecolor='black')

plt.title('Distribuzione delle Altezze nella Popolazione')
plt.xlabel('Altezza (cm)')
plt.ylabel('Frequenza')
plt.grid(True)
plt.show()
```


### Grafico a scatola (boxplot)



I boxplot (o diagrammi a scatola e baffi) sono grafici statistici che forniscono una rappresentazione sintetica della distribuzione dei dati numerici attraverso cinque statistiche principali, consentendo di identificare la presenza di outlier e di valutare la simmetria della distribuzione. 

Utilizzati per visualizzare la distribuzione di una variabile continua e per identificare i quartili, la mediana e i valori anomali (outliers).

Esempi d'uso:
* Distribuzione degli stipendi in diverse aziende: Utilizzato nelle risorse umane
* Distribuzione dei tempi di consegna per diversi corrieri: Utilizzato in logistica
* Prestazioni degli studenti in diverse materie: Utilizzato in educazionene


```python
# Generazione di dati casuali
np.random.seed(0)
data = np.random.randn(100)  # 100 valori casuali da una distribuzione normale
#print(data)

# Calcolo dei quartili
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)  # Mediana
q3 = np.percentile(data, 75)
iqr = q3 - q1
print(q1, q2, q3)

# Calcolare i limiti dei baffi
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print('\nbaffo inferiore', lower_bound)
print('\nbaffo superiore', upper_bound)

# Creazione del boxplot
plt.figure(figsize=(10, 6))

sns.boxplot(data=data)

plt.title('Boxplot di un singolo insieme di dati')
plt.ylabel('Valori')
plt.show()

```

```python
# Dati di esempio: distribuzione degli stipendi in diverse aziende
data = pd.DataFrame({
    'Azienda': ['Tech Solutions', 'Tech Solutions', 'Digital Services', 'Digital Services', 'Digital Services',
                'Media Corp', 'Media Corp', 'Media Corp', 'Media Corp', 'Media Corp'],
    'Stipendio': [40000, 42000, 38000, 39000, 41000, 35000, 36000, 37000, 38000, 20000]
    #'Stipendio': [40000, 42000, 38000, 39000, 41000, 35000, 36000, 37000, 38000, 70000]  #media corp 
})

# Calcolo dei quartili per ogni azienda
quartiles = data.groupby('Azienda')['Stipendio'].quantile([0.25, 0.5, 0.75]).unstack()
print(quartiles)

q1 = quartiles[0.25]
q3 = quartiles[0.75]
iqr = q3 - q1
print('Intervallo Interquartile', iqr)

# Calcolare i limiti dei baffi
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print('\nbaffo inferiore', lower_bound)
print('\nbaffo superiore', upper_bound)

# Trovare outlier per ogni azienda
outliers = {}
for azienda in data['Azienda'].unique(): #cicla sui valori unici presenti nella colonna 'Azienda' del DataFrame data. 
    azienda_data = data[data['Azienda'] == azienda]['Stipendio'] #filtra il DataFrame data per selezionare solo le righe relative all'azienda corrente.
    outliers[azienda] = azienda_data[(azienda_data < lower_bound[azienda]) | (azienda_data > upper_bound[azienda])]

print(outliers)

# Definizione di una palette di colori personalizzata 
palette_colors = {"Tech Solutions": "skyblue", "Digital Services": "salmon", "Media Corp": "gold"}
# Creazione del boxplot con Seaborn
plt.figure(figsize=(10, 6))

sns.boxplot(x='Azienda', y='Stipendio', hue='Azienda', data=data, palette=palette_colors)

plt.title('Distribuzione degli Stipendi per Azienda')
plt.xlabel('Azienda')
plt.ylabel('Stipendio ($)')
plt.show()
```


### Grafico a violino (violinplot) - Per visualizzare la distribuzione della densitÃ  di una variabile continua.

I grafici a violino sono strumenti utili per eseguire un'analisi esplorativa iniziale delle variabili in un set di dati. Essi facilitano la comprensione e il confronto delle distribuzioni delle variabili tra diverse categorie.

Sono particolarmente indicati quando il numero di osservazioni Ã¨ elevato. In situazioni con un grande numero di dati, ad esempio, l'utilizzo di grafici a dispersione (scatter plot) diventa impraticabile..


Esempi d'uso:
- Distribuzione dei punteggi degli esami di matematica per studenti di diversi gradi scolastici
- Distribuzione delle ore di sonno tra diverse fasce d'etÃ : Utilizzato in studi di salute per analizzare i pattern di sonno.
- Distribuzione delle valutazioni dei film per diversi generi: Utilizzato in analisi dei dati per confrontare la popolaritÃ  dei generi cinematografici.


```python
# Dati di esempio: distribuzione delle valutazioni dei film per diversi generi
data = pd.DataFrame({
    'Genere': ['Azione', 'Azione', 'Drammatico', 'Drammatico', 'Commedia', 'Commedia', 'Commedia', 'Horror', 'Horror', 'Horror'],
    'Valutazione': [7.5, 8.2, 6.9, 7.3, 6.8, 7.1, 7.0, 5.5, 6.2, 5.8]
})

# Creazione del violinplot con Seaborn
plt.figure(figsize=(10, 6))

sns.violinplot(x='Genere', y='Valutazione', data=data)

plt.title('Distribuzione delle Valutazioni per Genere di Film')
plt.xlabel('Genere', labelpad=30)
plt.ylabel('Valutazione', labelpad=30)
plt.show()
```

### spiegazione del grafico

Ogni "violino" nel grafico rappresenta la distribuzione delle valutazioni per un genere specifico di film. La forma e le caratteristiche del violino forniscono diverse informazioni:

- Larghezza del violino: Indica la distribuzione delle valutazioni in quel genere. Un violino piÃ¹ largo indica una maggiore densitÃ  delle valutazioni in quel range di valori.

- Linea bianca al centro del violino: Rappresenta la mediana delle valutazioni per quel genere di film. Questo punto divide il violino in due parti uguali in termini di numero di valutazioni.

- Forma complessiva del violino: Riflette la distribuzione dei dati per quel genere di film. Un violino simmetrico suggerisce una distribuzione simmetrica delle valutazioni, mentre un violino asimmetrico o con rigonfiamenti indica una distribuzione piÃ¹ complessa dei dati, che puÃ² includere code lunghe, cluster o altre caratteristiche non uniformi.
