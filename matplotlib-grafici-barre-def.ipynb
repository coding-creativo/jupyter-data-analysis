# Matplotlib
Matplotlib è una libreria di visualizzazione di dati per il linguaggio di programmazione Python. È molto potente e offre una vasta gamma di grafici e plot per la creazione di visualizzazioni di dati di alta qualità. È utilizzato ampiamente in ambiti come la scienza dei dati, l'analisi statistica, l'ingegneria e molti altri settori che richiedono la rappresentazione visuale dei dati.

Matplotlib può creare tanti tipi di grafici, tra cui:

- ***Grafici a linee***: per rappresentare relazioni tra dati continui.
- ***Scatter plot***: per visualizzare punti di dati senza connessioni lineari.
- ***Grafici a torta***: per esprimere proporzioni o percentuali.
- ***Grafici a barre***: per confrontare quantità tra diverse categorie.
- ***Istogrammi***: per mostrare la distribuzione dei dati.


## BAR - GRAFICI A BARRE

Per creare un grafico a barre in Matplotlib, possiamo utilizzare la funzione ***plt.bar()***. 

Passi principali per creare un grafico a barre:

* Prepariamo i dati da visualizzare nel grafico a barre. Generalmente, due liste: una per le etichette delle barre (es. nomi di categorie) e una per i valori corrispondenti.

* Utilizziamo il metodo plt.bar() per creare il grafico a barre, passando le etichette e i valori come argomenti principali.

* Aggiungere etichette alle sezioni, un titolo al grafico, e una legenda se necessario.

* Utilizziamo plt.show() per visualizzare il grafico.


## esempi d'utilizzo
- Comparare vendite di prodotti diversi in un negozio.
- Mostrare il numero di visite a un sito web da parte di utenti diversi giorni della settimana.
- Rappresentare la distribuzione di voti o punteggi in una classe.
- Numero di clienti serviti ogni mese in un ristorante.
- Numero di vendite mensili di un prodotto.
- Confrontare le medie di altezza tra diverse classi di età.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

# Dati di esempio
squadre = ['Squadra A', 'Squadra B', 'Squadra C', 'Squadra D']
punti = [23, 17, 35, 29]

# Creazione del grafico a barre
plt.bar(squadre, punti)

# Aggiunta di etichette agli assi e un titolo
plt.xlabel('squadre')
plt.ylabel('punti')
plt.title('Punteggi squadre')

# Aggiunta della legenda (opzionale)
# plt.legend(['Valori'])

# Visualizzazione del grafico
plt.show()
```

```python
#colori diversi
plt.bar(squadre, punti, color=['blue', 'green', 'red', 'purple'])

plt.show()
```

```python
#Larghezza delle barre: 
plt.bar(squadre, punti, width=0.5)
plt.show()
```

```python
## orietnamento orizzontale
# Creazione del grafico a barre
plt.barh(squadre, punti)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dati di esempio: nomi degli studenti e i loro voti
studenti = ['Mario', 'Giulia', 'Luca', 'Anna', 'Sara']
voti_mate = [85, 90, 70, 88, 92]
voti_ita = [78, 85, 65, 90, 87]

# Creazione del dataframe con Pandas
df = pd.DataFrame({
    'Studente': studenti,
    'Matematica': voti_mate,
    'Italiano': voti_ita,
})

#print(df)

# Melt del dataframe per ottenere un formato lungo per Seaborn 
# trasformiamo il DataFrame in un formato lungo, in modo che i voti in Matematica 
# e Italiano siano nella stessa colonna con un'altra colonna (Voto) che identifica il tipo di voto.
'''
Formato Largo: Le colonne corrispondono a diverse variabili o categorie. 
Ad esempio, nel DataFrame originale ogni studente ha una colonna separata per il voto in Matematica e Italiano.

Formato Lungo: Le colonne corrispondono a variabili categoriche o fattori che descrivono le osservazioni. 
In questo formato, ci sono meno colonne, ma più righe, ciascuna delle quali rappresenta un'osservazione singola con una
variabile categorica e il suo valore.

'''
df_melted = df.melt(id_vars=['Studente'], var_name='Materia', value_name='Voto')
print(df_melted)

# Creazione del grafico a barre con Seaborn
plt.figure(figsize=(10, 6))

#con seaborn prendo barplot
sns.barplot(x='Studente', y='Voto', hue='Materia', data=df_melted, palette='muted')

# Aggiunta di titolo e etichette degli assi
plt.title('Voti degli Studenti in Matematica e Italiano')
plt.xlabel('Studenti', labelpad=30)
plt.ylabel('Voto')
plt.legend(title='Materia')

# Mostra il grafico
plt.grid(True, axis='y')  # Abilita le linee di griglia sull'asse Y
plt.tight_layout()  # Ottimizza il layout per evitare sovrapposizioni
plt.show()
```
