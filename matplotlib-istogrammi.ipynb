Gli istogrammi vengono utilizzati in Matplotlib e in altre librerie di visualizzazione dei dati per rappresentare la distribuzione di frequenza di una variabile numerica continua. 

Alcuni esempi d'uso:
- la distribuzione delle altezze di una popolazione
- la distribuzione di caratteristiche demografiche come l'età della popolazione, il reddito familiare o il livello di istruzione
- la distribuzione dei tempi di attesa dei clienti in un negozio.
- la distribuzione dei rendimenti giornalieri di un'azione o di un indice di mercato
- la distribuzione delle temperature medie mensili o annuali, delle precipitazioni o di altri parametri climatici in diverse regioni.
- la distribuzione di parametri vitali come la pressione sanguigna, la frequenza cardiaca o i livelli di glucosio nel sangue



Confrontare distribuzioni: È possibile sovrapporre o affiancare più istogrammi per confrontare la distribuzione di diverse variabili o sottoinsiemi di dati.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generazione di dati casuali per i punteggi di matematica degli studenti
np.random.seed(13)
punteggi_matematica = np.random.randint(50, 100, size=100)  # Genera 100 punteggi casuali tra 50 e 100

# Creazione dell'istogramma
plt.figure(figsize=(10, 6))

plt.hist(punteggi_matematica, bins=10, edgecolor='black', alpha=0.7)

# Personalizzazione del grafico
plt.title('Distribuzione dei punteggi di matematica')
plt.xlabel('Punteggio di matematica')
plt.ylabel('Frequenza')
plt.grid(True, linestyle='--', alpha=0.5)

# Mostra il grafico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generazione di altezze casuali in cm
np.random.seed(0)
altezze = np.random.normal(loc=170, scale=10, size=200)

# Creazione dell'istogramma
plt.figure(figsize=(10, 6))

plt.hist(altezze, bins=20, edgecolor='black', alpha=0.7)

# Aggiunta di titolo e etichette degli assi
plt.title('Distribuzione delle altezze della popolazione')
plt.xlabel('Altezza (cm)')
plt.ylabel('Frequenza')

# Mostra il grafico
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
