
# Lezione su Seaborn

Seaborn è una libreria di visualizzazione dei dati basata su Matplotlib che fornisce un'interfaccia di alto livello per disegnare grafici statistici attraenti e informativi.



## Importazione della libreria

Iniziamo importando Seaborn e altre librerie necessarie.


```python

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Impostiamo uno stile di default per i grafici
sns.set(style="darkgrid")
```


## Grafici di base

### Grafico a dispersione (scatterplot)

Casi d'utilizzo:
* Relazione tra altezza e peso di un gruppo di persone. * Relazione tra spese pubblicitarie e vendite: Utilizzato in marketing per vedere se l'aumento della spesa pubblicitaria si traduce in maggiori vendite.* 
Relazione tra temperatura e consumo energeticoe.


```python

# Generazione di dati casuali per altezza e peso
np.random.seed(0)
altezza = np.random.normal(loc=170, scale=10, size=100)
peso = altezza * 0.6 + np.random.normal(loc=0, scale=7, size=100)

# Creazione del grafico a dispersione
plt.figure(figsize=(10, 6))
sns.scatterplot(x=altezza, y=peso, color='skyblue',  alpha=0.8)
plt.title('Relazione tra Altezza e Peso')
plt.xlabel('Altezza (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)
plt.show()
```


### Grafico a barre 

Per confrontare quantità tra diverse categorie

Esempi d'utilizzo:
* Confronto tra le vendite mensili di diversi prodotti* Numero medio di ore di studio settimanali per studenti di diverse facolt
* Distribuzione della popolazione per regione




Creiamo un grafico a barre utilizzando un DataFrame.


```python

# Dati categoriali
data = pd.DataFrame({
    'Categoria': ['A', 'B', 'C', 'D', 'E'],
    'Valore': [5, 7, 3, 8, 4]
})


# Creazione del grafico
plt.figure(figsize=(10, 6))
sns.barplot(x='Categoria', y='Valore', data=data)
plt.title('Grafico a barre')
plt.xlabel('Categoria')
plt.ylabel('Valore')
plt.show()
```


## Visualizzazione delle distribuzioni

## histplot e kde 

Esempi d'uso:

* Distribuzione delle altezze in una popolazione: Utilizzato in antropologia e biologia per analizzare la variabilità delle altezze.
* Distribuzione delle età dei clienti di un negozio: Utilizzato nel marketing per capire la composizione demografica dei clienti.
* Distribuzione dei tempi di risposta in un test psicologico: Utilizzato in psicologia per analizzare la rapidità di risposta..


```python
# Dati di esempio: età dei clienti
np.random.seed(0)
eta_clienti = np.random.randint(18, 70, size=200)  # Generazione di 200 età casuali comprese tra 18 e 70 anni
#print(eta_clienti)

# Creazione dell'istogramma
plt.figure(figsize=(10, 6))
sns.histplot(eta_clienti, bins=15, color='skyblue', edgecolor='black')
plt.title('Distribuzione delle Età dei Clienti')
plt.xlabel('Età')
plt.ylabel('Frequenza')
plt.grid(True)  # Aggiunge una griglia di riferimento al grafico
plt.show()
```


### kde
kde, acronimo di Kernel Density Estimation, è una tecnica statistica utilizzata per stimare la densità di probabilità di una variabile casuale. Nei grafici statistici, come gli istogrammi o i violin plot, il KDE è spesso utilizzato per fornire una rappresentazione liscia e continua della distribuzione dei dati.

kde cerca di modellare la distribuzione sottostante dei dati, fornendo una rappresentazione più dettagliata rispetto agli istogrammimi

```python
# Generazione di dati casuali per le altezze (simulazione di una distribuzione normale)
np.random.seed(0)
altezze = np.random.normal(loc=170, scale=10, size=1000)  # Media 170 cm, deviazione standard 10 cm, 1000 campioni
#print(altezze)
# Creazione del grafico con histplot e KDE sovrapposto
plt.figure(figsize=(10, 6))
sns.histplot(altezze, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribuzione delle Altezze nella Popolazione')
plt.xlabel('Altezza (cm)')
plt.ylabel('Frequenza')
plt.grid(True)
plt.show()
```


### Grafico a scatola (boxplot)



I boxplot (o diagrammi a scatola e baffi) sono grafici statistici che forniscono una rappresentazione sintetica della distribuzione dei dati numerici attraverso cinque statistiche principali, consentendo di identificare la presenza di outlier e di valutare la simmetria della distribuzione. 

Utilizzati per visualizzare la distribuzione di una variabile continua e per identificare i quartili, la mediana e i valori anomali (outliers).

Esempi d'uso:
* Distribuzione degli stipendi in diverse aziende: Utilizzato nelle risorse umane
* Distribuzione dei tempi di consegna per diversi corrieri: Utilizzato in logistica
* Prestazioni degli studenti in diverse materie: Utilizzato in educazionene


```python
# Generazione di dati casuali
np.random.seed(0)
data = np.random.randn(100)  # 100 valori casuali da una distribuzione normale
#print(data)

# Calcolo dei quartili
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)  # Mediana
q3 = np.percentile(data, 75)
print(q1, q2, q3)

# Calcolare i limiti dei baffi
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print('\nbaffo inferiore', lower_bound)
print('\nbaffo superiore', upper_bound)

# Creazione del boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Boxplot di un singolo insieme di dati')
plt.ylabel('Valori')
plt.show()

```

```python
# Dati di esempio: distribuzione degli stipendi in diverse aziende
data = pd.DataFrame({
    'Azienda': ['Tech Solutions', 'Tech Solutions', 'Digital Services', 'Digital Services', 'Digital Services',
                'Media Corp', 'Media Corp', 'Media Corp', 'Media Corp', 'Media Corp'],
    'Stipendio': [40000, 42000, 38000, 39000, 41000, 35000, 36000, 37000, 38000, 39000]
    #'Stipendio': [40000, 42000, 38000, 39000, 41000, 35000, 36000, 37000, 38000, 70000]  #media corp 
})

# Calcolo dei quartili per ogni azienda
quartiles = data.groupby('Azienda')['Stipendio'].quantile([0.25, 0.5, 0.75]).unstack()
print(quartiles)

q1 = quartiles[0.25]
q3 = quartiles[0.75]
iqr = q3 - q1
print('Intervallo Interquartile', iqr)

# Calcolare i limiti dei baffi
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print('\nbaffo inferiore', lower_bound)
print('\nbaffo superiore', upper_bound)

# Trovare outlier per ogni azienda
outliers = {}
for azienda in data['Azienda'].unique(): #cicla sui valori unici presenti nella colonna 'Azienda' del DataFrame data. 
    azienda_data = data[data['Azienda'] == azienda]['Stipendio'] #filtra il DataFrame data per selezionare solo le righe relative all'azienda corrente.
    outliers[azienda] = azienda_data[(azienda_data < lower_bound[azienda]) | (azienda_data > upper_bound[azienda])]

#print(outliers)

# Definizione di una palette di colori personalizzata 
palette_colors = {"Tech Solutions": "skyblue", "Digital Services": "salmon", "Media Corp": "gold"}
# Creazione del boxplot con Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Azienda', y='Stipendio', hue='Azienda', data=data, palette=palette_colors)
plt.title('Distribuzione degli Stipendi per Azienda')
plt.xlabel('Azienda')
plt.ylabel('Stipendio ($)')
plt.show()
```


### Grafico a violino (violinplot) - Per visualizzare la distribuzione della densità di una variabile continua.

I grafici a violino sono strumenti utili per eseguire un'analisi esplorativa iniziale delle variabili in un set di dati. Essi facilitano la comprensione e il confronto delle distribuzioni delle variabili tra diverse categorie.

Sono particolarmente indicati quando il numero di osservazioni è elevato. In situazioni con un grande numero di dati, ad esempio, l'utilizzo di grafici a dispersione (scatter plot) diventa impraticabile..


Esempi d'uso:
- Distribuzione dei punteggi degli esami di matematica per studenti di diversi gradi scolastici
- Distribuzione delle ore di sonno tra diverse fasce d'età: Utilizzato in studi di salute per analizzare i pattern di sonno.
- Distribuzione delle valutazioni dei film per diversi generi: Utilizzato in analisi dei dati per confrontare la popolarità dei generi cinematografici.


```python
# Dati di esempio: distribuzione delle valutazioni dei film per diversi generi
data = pd.DataFrame({
    'Genere': ['Azione', 'Azione', 'Drammatico', 'Drammatico', 'Commedia', 'Commedia', 'Commedia', 'Horror', 'Horror', 'Horror'],
    'Valutazione': [7.5, 8.2, 6.9, 7.3, 6.8, 7.1, 7.0, 5.5, 6.2, 5.8]
})

# Creazione del violinplot con Seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='Genere', y='Valutazione', data=data)
plt.title('Distribuzione delle Valutazioni per Genere di Film')
plt.xlabel('Genere', labelpad=30)
plt.ylabel('Valutazione', labelpad=30)
plt.show()
```

### spiegazione del grafico

Ogni "violino" nel grafico rappresenta la distribuzione delle valutazioni per un genere specifico di film. La forma e le caratteristiche del violino forniscono diverse informazioni:

- Larghezza del violino: Indica la distribuzione delle valutazioni in quel genere. Un violino più largo indica una maggiore densità delle valutazioni in quel range di valori.

- Linea bianca al centro del violino: Rappresenta la mediana delle valutazioni per quel genere di film. Questo punto divide il violino in due parti uguali in termini di numero di valutazioni.

- Forma complessiva del violino: Riflette la distribuzione dei dati per quel genere di film. Un violino simmetrico suggerisce una distribuzione simmetrica delle valutazioni, mentre un violino asimmetrico o con rigonfiamenti indica una distribuzione più complessa dei dati, che può includere code lunghe, cluster o altre caratteristiche non uniformi.

## DATASET
Seaborn fornisce alcuni dataset integrati che possono essere caricati direttamente per scopi di esplorazione e analisi. Questi dataset sono utilizzati comunemente per esempi didattici e per testare le funzionalità di Seaborn. Per caricare un dataset integrato di Seaborn, possiamo utilizzare la funzione load_dataset()

Alcuni dataset:
- iris: Dataset con misurazioni di lunghezza e larghezza del sepalo e del petalo di tre specie di iris
- tips: Dataset con dati su conti e mance in un ristorante.
- titanic: Dataset con informazioni dettagliate su passeggeri a bordo del Titanic, inclusi dati come la classe del biglietto, il sesso, l'età e se il passeggero ha sopravvissuto.
- flights: Dataset con dati mensili sui passeggeri dei voli di una compagnia aerea.
- diamonds: Dataset con dati sui diamanti, includendo carati, colore, chiarezza e prezzo.
- planets: Dataset con dati su pianeti extrasolari scoperti, inclusi caratteristiche come massa, raggio e metodo di scoperta.

```python

# Dati casuali
data = sns.load_dataset('titanic')
print(data)

# Creazione del grafico
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', data=data)
plt.title('Count plot')
plt.xlabel('Classe')
plt.ylabel('Conteggio')
plt.show()
```


### Pairplot

Creiamo un pairplot per visualizzare le relazioni tra tutte le variabili in un DataFrame.


```python

# Dati casuali
data = sns.load_dataset('iris')

# Creazione del grafico
sns.pairplot(data, hue='species')
plt.show()
```


### Point plot

Creiamo un point plot per visualizzare la media di una variabile categoriale.


```python

# Creazione del grafico
plt.figure(figsize=(10, 6))
sns.pointplot(x='class', y='age', data=data, hue='sex', dodge=True, markers=['o', 's'], linestyles=['-', '--'])
plt.title('Point plot')
plt.xlabel('Classe')
plt.ylabel('Età media')
plt.show()
```


## Personalizzazione e temi

Seaborn offre diversi temi per personalizzare l'aspetto dei grafici.


```python

# Dati casuali
data = np.random.randn(1000)

# Creazione del grafico con un tema personalizzato
sns.set_theme(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='m')
plt.title('Istogramma con tema personalizzato')
plt.xlabel('Valore')
plt.ylabel('Frequenza')
plt.show()
```


## Integrazione con Matplotlib

Seaborn è costruito su Matplotlib, quindi possiamo utilizzare le funzioni di Matplotlib per personalizzare ulteriormente i nostri grafici.


```python

# Dati casuali
data = sns.load_dataset('iris')

# Creazione del grafico
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Scatter plot con personalizzazione Matplotlib')
plt.xlabel('Lunghezza sepalo')
plt.ylabel('Larghezza sepalo')

# Personalizzazione con Matplotlib
plt.axhline(y=3, color='r', linestyle='--')
plt.axvline(x=5, color='b', linestyle='--')
plt.show()
```
