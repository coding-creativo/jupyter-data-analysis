# Matplotlib
Matplotlib è una libreria di visualizzazione di dati per il linguaggio di programmazione Python. È molto potente e offre una vasta gamma di grafici e plot per la creazione di visualizzazioni di dati di alta qualità. È utilizzato ampiamente in ambiti come la scienza dei dati, l'analisi statistica, l'ingegneria e molti altri settori che richiedono la rappresentazione visuale dei dati.

Matplotlib può creare tanti tipi di grafici, tra cui:

- ***Grafici a linee***: per rappresentare relazioni tra dati continui.
- ***Scatter plot***: per visualizzare punti di dati senza connessioni lineari.
- ***Grafici a torta***: per esprimere proporzioni o percentuali.
- ***Grafici a barre***: per confrontare quantità tra diverse categorie.
- ***Istogrammi***: per mostrare la distribuzione dei dati.


## Grafico a Linee

Per creare un grafico a linee in Matplotlib, possiamo utilizzare la funzione ***plt.plot()***. 

Passi principali per creare un grafico a linee:

* Definiamo i dati: Prepariamo i dati da visualizzare come linee. Di soluto due array o liste, uno per l'asse X e uno per l'asse Y, che rappresentano le coordinate dei punti nel grafico.

* Chiamata a plt.plot(): Utilizziamo plt.plot() passando i dati sull'asse X e sull'asse Y come argomenti principali. Si possono anche specificare opzioni aggiuntive come il colore della linea, lo stile della linea, i marcatori e tanto altro.

* Personalizzazione del grafico: Aggiungiamo etichette agli assi, un titolo al grafico, una legenda se necessario e abilitiamo la griglia per migliorare la leggibilità del grafico.

* Mostriamo il grafico: Utilizziamo plt.show() per visualizzare il grafico.

```python
import matplotlib.pyplot as plt
```

## ESEMPIO 1

```python
# 1) GRAFICO A LINEE
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creazione del grafico a linee
plt.figure(figsize=(8, 4))  
''' Definizione delle dimensioni della figura - figure è un parametro opzionale della funzione plt.figure() che specifica le dimensioni 
    della figura in pollici. Nel caso specifico, (8, 4) indica che la larghezza della figura sarà di 8 pollici e l'altezza sarà di 4 pollici.
'''
plt.plot(x, y, marker='s', linestyle='-', color='y', label='Dati')  
''' Creiamo un grafico a linee che rappresenta i dati forniti dalle liste x e y 
    Specifichiamo:
        1) Il tipo di marcatore da utilizzare sui punti del grafico. 
            In questo caso, 'o' indica che si vogliono i cerchi come marcatori per ciascun punto.
            'o': Cerchio, 's': Quadrato, '^': Triangolo verso l'alto, 'v': Triangolo verso il basso, '>': Triangolo verso destra, 
            '<': Triangolo verso sinistra, 'x': X, '+': Più, 'D': Diamante, 'p': Pentagono, '*': Stella, 
        2) linestyle='-'. Specifica lo stile della linea utilizzata per connettere i punti. 
            '-' indica una linea continua; '--' Linea tratteggiata; ':' Linea a puntini
        3) color='b' nomi comuni (come 'b' per blu, 'r' per rosso, ecc.) o codici esadecimali ('#FFA500' per arancione).
        4) label='Dati': Questo è il nome che verrà utilizzato per identificare la serie di dati nel caso in cui si desideri 
            aggiungere una legenda al grafico.
'''
plt.xlabel('X')  # Etichetta asse X
plt.ylabel('Y')  # Etichetta asse Y
plt.title('Grafico a Linee')  # Titolo del grafico
plt.legend()  # Mostra la legenda
plt.grid(True)  # Mostra la griglia
plt.show()  # Mostra il grafico
```

#### STESSO ESEMPIO CON NUMPY

```python
#2) STESSO ESEMPIO CON NUMPY
import numpy as np
x = np.arange(0,5) #valori da 0 a 4
y = np.arange(0,10,2) #valori 0, 2, 4, 6, 8

# Creazione del grafico a linee
plt.figure(figsize=(8, 4))  # Definizione delle dimensioni della figura
plt.plot(x, y, marker='o', linestyle='-', linewidth=2, markersize=10,  color='y', markerfacecolor='yellow', markeredgecolor='black', markeredgewidth=2, alpha=0.7, label='Dati') 
''' linewidth (o lw): Imposta lo spessore della linea che connette i punti nel grafico.
    Possibili valori: Numeri positivi che rappresentano lo spessore della linea in punti (pixels).
    Esempio: linewidth=2 imposta una linea con uno spessore di 2 punti.

    markersize (o ms): Determina la dimensione dei marcatori che rappresentano i punti nel grafico.
    Possibili valori: Numeri positivi che indicano il diametro del marcatore in punti.
    Esempio: markersize=10 imposta marcatori con un diametro di 10 punti.

    alpha: Regola la trasparenza dei marcatori e della linea nel grafico.
    Possibili valori: Numeri compresi tra 0 (completamente trasparente) e 1 (completamente opaco).
    Esempio: alpha=0.7 imposta una trasparenza del 70% per sia i marcatori che la linea.
    
    markerfacecolor (o mfc): Specifica il colore di riempimento dei marcatori nel grafico.
    Possibili valori: Stringhe che rappresentano nomi di colori (come 'yellow', 'blue', ecc.) o codici esadecimali.
    Esempio: markerfacecolor='yellow' riempie i marcatori con il colore giallo.

    markeredgecolor (o mec): Imposta il colore del bordo dei marcatori nel grafico.
    Possibili valori: Stringhe che rappresentano nomi di colori (come 'black', 'red', ecc.) o codici esadecimali.
    Esempio: markeredgecolor='black' imposta il bordo dei marcatori di colore nero.
    
    markeredgewidth (o mew): Controlla lo spessore del bordo dei marcatori nel grafico.
    Possibili valori: Numeri positivi che rappresentano lo spessore del bordo in punti.
    Esempio: markeredgewidth=2 imposta lo spessore del bordo dei marcatori a 2 punti.
'''
plt.xlabel('X', fontsize=12, fontstyle='italic')  # Etichetta asse X
plt.ylabel('Y', fontsize=12, fontweight='bold')  # Etichetta asse Y
plt.title('Grafico a Linee',fontsize=22)  # Titolo del grafico
plt.legend()  # Mostra la legenda
plt.grid(True)  # Mostra la griglia
plt.show()  # Mostra il grafico
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creiamo un dataframe di esempio
np.random.seed(13)  # Per riproducibilità
dates = pd.date_range(start='2023-01-01', periods=20)
cities = ['Roma', 'Milano', 'Torino']
data = {
    'Date': np.repeat(dates, len(cities)),
    'City': np.tile(cities, len(dates)),
    'Temperature': np.random.randint(low=0, high=30, size=len(dates) * len(cities))
}

df = pd.DataFrame(data)

# Visualizziamo le prime righe del dataframe
print(df.head())
```

```python
# Creazione del grafico a linee per l'andamento delle temperature nelle città
plt.figure(figsize=(12, 8))

# Ciclo per creare una linea per ogni città
for city in cities:
    city_data = df[df['City'] == city]
    plt.plot(city_data['Date'], city_data['Temperature'], marker='o', linestyle='-', label=city)

# Personalizzazione del grafico
plt.title('Andamento delle temperature nelle città')
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.grid(True)
plt.legend()

# Mostra il grafico
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creiamo un dataframe di esempio
np.random.seed(0)  # Per riproducibilità
heights = np.random.normal(loc=68, scale=3, size=100)
weights = heights * 2.5 + np.random.normal(loc=0, scale=5, size=100)  # Relazione fittizia

data = {
    'Height (inches)': heights,
    'Weight (pounds)': weights
}

df = pd.DataFrame(data)

# Visualizziamo le prime righe del dataframe
print(df.head())
```

```python
# Creazione dello scatter plot per la relazione tra altezza e peso
plt.figure(figsize=(10, 6))

plt.scatter(df['Height (inches)'], df['Weight (pounds)'], color='blue', alpha=0.5)
plt.title('Scatter plot: Altezza vs Peso')
plt.xlabel('Altezza (pollici)')
plt.ylabel('Peso (libbre)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## GRAFICO A TORTA


```python
# Dati per il grafico a torta
labels = np.array(['A', 'B', 'C', 'D'])
sizes = np.array([15, 30, 45, 10]) # Percentuali
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Esplosione del primo cuneo

# Crea il grafico a torta
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equalizza l'asse per renderlo un cerchio

plt.title('Esempio di grafico a torta')
plt.show()
```

```python
# Dati di esempio
data = {'Categoria': ['Elettronica', 'Abbigliamento', 'Alimentari', 'Libri', 'Sport'],
        'Vendite (%)': [25, 20, 30, 15, 10]}

sales_data = pd.DataFrame(data)

# Estraiamo i dati per il grafico a torta
labels = sales_data['Categoria']
sizes = sales_data['Vendite (%)']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']
explode = (0.1, 0, 0, 0, 0)  # Esplosione del primo cuneo

# Crea il grafico a torta
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equalizza l'asse per renderlo un cerchio

plt.title('Distribuzione percentuale delle vendite per categoria')
plt.show()
```

## GRAFICO A BARRE

```python
import matplotlib.pyplot as plt

# Dati di esempio: vendite mensili per quattro prodotti (A, B, C, D)
mesi = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile', 'Maggio']
vendite_prodotto_A = [15000, 18000, 21000, 19000, 20000]
vendite_prodotto_B = [12000, 15000, 17000, 16000, 18000]
vendite_prodotto_C = [10000, 11000, 13000, 12000, 14000]
vendite_prodotto_D = [9000, 9500, 11000, 10500, 12000]

# Creazione del grafico a barre verticale per confrontare le vendite mensili dei prodotti
plt.figure(figsize=(10, 6))  # Imposta le dimensioni della figura (larghezza, altezza)
larghezza_barre = 0.2  # Larghezza delle singole barre

# Posizioni delle barre per ogni mese
posizioni_mesi = list(range(len(mesi)))

# Tracciamento delle barre per ogni prodotto
plt.bar([pos for pos in posizioni_mesi], vendite_prodotto_A, width=larghezza_barre, color='blue', label='Prodotto A')
plt.bar([pos for pos in posizioni_mesi], vendite_prodotto_B, width=larghezza_barre, color='orange', label='Prodotto B')
plt.bar([pos for pos in posizioni_mesi], vendite_prodotto_C, width=larghezza_barre, color='green', label='Prodotto C')
plt.bar([pos for pos in posizioni_mesi], vendite_prodotto_D, width=larghezza_barre, color='red', label='Prodotto D')

# Etichette degli assi e titolo
plt.xlabel('Mesi')
plt.ylabel('Vendite')
plt.title('Vendite Mensili per Prodotto')
plt.xticks(posizioni_mesi, mesi)  # Etichette degli assi X

# Mostra la legenda
plt.legend()

# Mostra il grafico
plt.grid(True, axis='y')  # Abilita le linee di griglia sull'asse Y
plt.tight_layout()  # Ottimizza il layout per evitare sovrapposizioni
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Dati di esempio: vendite mensili per quattro prodotti (A, B, C, D)
mesi = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile', 'Maggio']
vendite_prodotto_A = [15000, 18000, 21000, 19000, 20000]
vendite_prodotto_B = [12000, 15000, 17000, 16000, 18000]
vendite_prodotto_C = [10000, 11000, 13000, 12000, 14000]
vendite_prodotto_D = [9000, 9500, 11000, 10500, 12000]

# Creazione del dataframe per Seaborn
import pandas as pd
df = pd.DataFrame({
    'Mese': mesi,
    'Prodotto A': vendite_prodotto_A,
    'Prodotto B': vendite_prodotto_B,
    'Prodotto C': vendite_prodotto_C,
    'Prodotto D': vendite_prodotto_D
})

# Impostazioni grafiche di Seaborn
plt.figure(figsize=(10, 6))  # Imposta le dimensioni della figura (larghezza, altezza)

# Tracciamento del grafico a barre con Seaborn
sns.barplot(x='Mese', y='value', hue='variable', data=pd.melt(df, ['Mese']), palette='muted')

# Etichette degli assi e titolo
plt.xlabel('Mesi')
plt.ylabel('Vendite')
plt.title('Vendite Mensili per Prodotto')

# Mostra la legenda
plt.legend(title='Prodotto', loc='upper left')

# Mostra il grafico
plt.grid(True, axis='y')  # Abilita le linee di griglia sull'asse Y
plt.tight_layout()  # Ottimizza il layout per evitare sovrapposizioni
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dati di esempio: vendite mensili per quattro prodotti (A, B, C, D)
mesi = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile', 'Maggio']
vendite_prodotto_A = [15000, 18000, 21000, 19000, 20000]
vendite_prodotto_B = [12000, 15000, 17000, 16000, 18000]
vendite_prodotto_C = [10000, 11000, 13000, 12000, 14000]
vendite_prodotto_D = [9000, 9500, 11000, 10500, 12000]

# Creazione del grafico a barre verticale per confrontare le vendite mensili dei prodotti
plt.figure(figsize=(10, 6))  # Imposta le dimensioni della figura (larghezza, altezza)
larghezza_barre = 0.2  # Larghezza delle singole barre

# Posizioni delle barre per ogni mese
posizioni_mesi = list(range(len(mesi)))

# Tracciamento delle barre per ogni prodotto
plt.bar([pos - larghezza_barre for pos in posizioni_mesi], vendite_prodotto_A, width=larghezza_barre, color='blue', label='Prodotto A')
plt.bar(posizioni_mesi, vendite_prodotto_B, width=larghezza_barre, color='orange', label='Prodotto B')
plt.bar([pos + larghezza_barre for pos in posizioni_mesi], vendite_prodotto_C, width=larghezza_barre, color='green', label='Prodotto C')
plt.bar([pos + 2*larghezza_barre for pos in posizioni_mesi], vendite_prodotto_D, width=larghezza_barre, color='red', label='Prodotto D')

# Etichette degli assi e titolo
plt.xlabel('Mesi')
plt.ylabel('Vendite')
plt.title('Vendite Mensili per Prodotto')
plt.xticks(posizioni_mesi, mesi)  # Etichette degli assi X

# Mostra la legenda
plt.legend()

# Mostra il grafico
plt.grid(True, axis='y')  # Abilita le linee di griglia sull'asse Y
plt.tight_layout()  # Ottimizza il layout per evitare sovrapposizioni
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dati di esempio: nomi degli studenti e i loro voti
studenti = ['Mario', 'Giulia', 'Luca', 'Anna', 'Sara']
voti_mate = [85, 90, 70, 88, 92]
voti_ita = [78, 85, 65, 90, 87]  # Voti degli studenti in Italiano

# Creazione del dataframe con Pandas
df = pd.DataFrame({
    'Studente': studenti * 2,  # Ripetiamo gli studenti per Matematica e Italiano
    'Voto': voti_mate + voti_ita,
    'Materia': ['Matematica'] * len(studenti) + ['Italiano'] * len(studenti)
})
df
```

```python
import seaborn as sns

# Impostazione dello stile di seaborn
sns.set(style='whitegrid')

# Creazione del grafico a barre con Seaborn
plt.figure(figsize=(10, 6))  # Imposta le dimensioni della figura (larghezza, altezza)
sns.barplot(x='Studente', y='Voto', hue='Materia', data=df, palette='pastel')

# Etichette degli assi e titolo
plt.xlabel('Studenti')
plt.ylabel('Voti')
plt.title('Voti degli Studenti per Materia')

# Mostra il grafico
plt.legend(title='Materia', loc='upper right')  # Mostra la legenda con titolo e posizione
plt.tight_layout()  # Ottimizza il layout per evitare sovrapposizioni
plt.show()
```

```python
# URL diretto al file CSV
url = 'https://raw.githubusercontent.com/ShapeLab/ZooidsCompositePhysicalizations/master/Zooid_Vis/bin/data/student-dataset.csv'

# Carica il DataFrame dal file CSV
df = pd.read_csv(url)

# Visualizza le prime righe del DataFrame per identificare le colonne
print(df.head())

# Seleziona le colonne di punteggio di interesse
math_scores = df['math.grade']
# Verifica il nome esatto della colonna per i punteggi di lettura
reading_scores = df['english.grade']  # Assicurati che questo nome sia corretto
writing_scores = df['sciences.grade']

# Crea tre subplot per gli istogrammi dei tre punteggi
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Istogramma per il punteggio di matematica
ax1.hist(math_scores, bins=20, edgecolor='black', alpha=0.7)
ax1.set_title('Distribuzione dei punteggi di matematica')
ax1.set_xlabel('Punteggio di matematica')
ax1.set_ylabel('Frequenza')
ax1.grid(True, linestyle='--', alpha=0.5)

# Istogramma per il punteggio di lettura
ax2.hist(reading_scores, bins=20, edgecolor='black', alpha=0.7)
ax2.set_title('Distribuzione dei punteggi di lettura')
ax2.set_xlabel('Punteggio di lettura')
ax2.set_ylabel('Frequenza')
ax2.grid(True, linestyle='--', alpha=0.5)

# Istogramma per il punteggio di scrittura
ax3.hist(writing_scores, bins=20, edgecolor='black', alpha=0.7)
ax3.set_title('Distribuzione dei punteggi di scrittura')
ax3.set_xlabel('Punteggio di scrittura')
ax3.set_ylabel('Frequenza')
ax3.grid(True, linestyle='--', alpha=0.5)

# Aggiusta il layout e mostra i grafici
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# URL diretto al file CSV
url = 'https://raw.githubusercontent.com/ShapeLab/ZooidsCompositePhysicalizations/master/Zooid_Vis/bin/data/student-dataset.csv'

# Carica il DataFrame dal file CSV
df = pd.read_csv(url)

# Seleziona le colonne di punteggio di interesse
math_scores = df['math.grade']
reading_scores = df['english.grade']
writing_scores = df['sciences.grade']

# Crea il grafico per i punteggi di matematica
plt.hist(math_scores, bins=20, edgecolor='black', alpha=0.7, label='Matematica', color='blue')

# Crea il grafico per i punteggi di lettura
plt.hist(reading_scores, bins=20, edgecolor='black', alpha=0.7, label='Lettura', color='green')

# Crea il grafico per i punteggi di scrittura
plt.hist(writing_scores, bins=20, edgecolor='black', alpha=0.7, label='Scrittura', color='orange')

# Aggiungi titolo e etichette degli assi
plt.title('Distribuzione dei punteggi di matematica, lettura e scrittura')
plt.xlabel('Punteggio')
plt.ylabel('Frequenza')
plt.grid(True, linestyle='--', alpha=0.5)

# Aggiungi la legenda
plt.legend()

# Mostra il grafico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creiamo un DataFrame di esempio
data = {
    'StudentID': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'MathScore': [85, 70, 92, 78, 88],
    'ReadingScore': [78, 82, 90, 85, 79],
    'WritingScore': [80, 75, 88, 84, 81]
}

df = pd.DataFrame(data)

# Numero di studenti
num_students = len(df)
# Larghezza delle barre
bar_width = 0.25

# Posizione delle barre sull'asse x
r1 = np.arange(num_students)
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Creiamo il plot
plt.figure(figsize=(10, 6))

plt.bar(r1, df['MathScore'], color='blue', width=bar_width, edgecolor='grey', label='Matematica')
plt.bar(r2, df['ReadingScore'], color='green', width=bar_width, edgecolor='grey', label='Lettura')
plt.bar(r3, df['WritingScore'], color='orange', width=bar_width, edgecolor='grey', label='Scrittura')

# Aggiungiamo le etichette sugli assi e il titolo
plt.xlabel('Studenti', fontweight='bold')
plt.ylabel('Punteggi', fontweight='bold')
plt.title('Punteggi di Matematica, Lettura e Scrittura per Studente', fontweight='bold')

# Aggiungiamo le etichette sugli assi x e il titolo
plt.xticks([r + bar_width for r in range(num_students)], df['StudentID'])
plt.legend()

# Mostrare il grafico
plt.show()
```
