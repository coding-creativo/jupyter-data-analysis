## LOC E ILOC

```python
df.head()
```

```python
# 12 Selezione di righe e colonne con gli indici
df.iloc[0, :] #seleziona la prima riga e tutte le colonne [0,:] ":" è implicito
```

```python
# 12.a
df.iloc[0:2] #seleziona le prime due righe del DataFrame df utilizzando l'indice di posizione (iloc)
```

```python
# 12.b
df.iloc[:, 4] #[row, col] -> tutte le righe e la colonna con indice 4
#da notare che non da le intestazioni perchè è una serie!
```

```python
# 12.c
# per avere un DataFrame possiamo convertire la Series e specificare i nomi delle colonne manualmente
column_values = df.iloc[:, 4]
column_df = pd.DataFrame(column_values, columns=['PizzaName'])
column_df
```

```python
# 13 - Selezionare in base ai nomi delle colonne
df.loc[0, 'PizzaName'] #Solo la prima riga, COLUMN= 'name'  #occhio è case sensitive!!!
```

```python
# 13.a - Selezionare in base ai nomi delle colonne
df.loc[10:20, ['PizzaName', 'type']] #selezionare più righe e più colonne
```

```python
# 13.b - Selezionare in base ai nomi delle colonne
df.loc[0,'PizzaName'] #ALL ROWS, COLUMN= 'PizzaName'
```

## DATI STATISTICI

```python
# 14 Riassunto statistico
df.describe()
```

```python
# 14.a Riassunto statistico
df['price'].describe()
```

```python
# 15 value_counts()
# Conta il numero di occorrenze di ogni tipo di pizza - ovvero le pizze vendute per ciascun nome - sono già ordinate in ordine decrescente
pizza_counts = df['PizzaName'].value_counts()
pizza_counts
```

```python
# 16 type è una funzione built-in di Python che restituisce il tipo di dato di un oggetto. 
type(pizza_counts)
```

```python
# 17 Accesso all'indice della prima riga
most_sold_pizza_name = pizza_counts.index[0]
most_sold_pizza_name
```

```python
# 18 - Trova la pizza con il maggior numero di vendite
most_sold_pizza = pizza_counts.iloc[0]
most_sold_pizza
```

```python
# 19 convertire una serie in un dataframe
pizza_counts_df = pizza_counts.to_frame(name="Conteggio_Pizze")
pizza_counts_df
```

## FILTRARE I DATI

```python
# 20 - Filtrare per tipo di pizza 'supreme'
supreme_pizzas = df[df['type'] == 'supreme']
supreme_pizzas
```

```python
# 21 Filtra per tipo di pizza 'supreme' e dimensione 'L'
supreme_large_pizzas = df[(df['type'] == 'supreme') & (df['size'] == 'L')]
supreme_large_pizzas
```

## OPERAZIONI COMUNI

```python
# 22 Calcola la media dei prezzi delle pizze
prezzo_medio = df['price'].mean()
prezzo_medio
```

```python
# 23 Calcola il prezzo massimo
prezzo_medio = df['price'].max()
prezzo_medio
```

```python
# 24 Calcola il prezzo minimo
prezzo_medio = df['price'].min()
prezzo_medio
```

```python
# 25 Calcola la deviazione standard
prezzo_medio = df['price'].std()
prezzo_medio
```
