```python
import pandas as pd
import matplotlib.pyplot as plt
# 1) Caricamento del dataset del Titanic da un file CSV
df = pd.read_csv('titanic.csv')
```

```python
# 2) Calcolo del numero totale di passeggeri nel dataset
numero_passeggeri = len(df)
numero_passeggeri
```

```python
# 3) Calcolo della tariffa media per classe
tariffa_media_per_classe = df.groupby('Pclass')['Fare'].mean()
print("\nTariffa media per classe:")
tariffa_media_per_classe
```

```python

# Grafico a linee
plt.figure(figsize=(10, 6))
plt.plot(tariffa_media_per_classe.index, tariffa_media_per_classe.values, marker='o', linestyle='-', color='b')
plt.title('Prezzo medio del biglietto nelle diverse classi')
plt.xlabel('Classe')
plt.ylabel('Prezzo medio del biglietto')
plt.grid(True)
plt.xticks(tariffa_media_per_classe.index, ['1a Classe', '2a Classe', '3a Classe'])
plt.show()
```

```python
# 4) Mostra le prime 5 righe del dataset
print("\nPrime 5 righe del dataset:")
df.head()
```

```python
# 5) Conteggio dei passeggeri per ciascun sesso
conteggio_per_sesso = df['Sex'].value_counts()
print("\nConteggio dei passeggeri per sesso:")
conteggio_per_sesso
```

```python
# 6) Mostra la colonna dei sopravvissuti
print("\nColonna dei sopravvissuti:")
df['Survived']
```

```python
# 7) Conta il numero di sopravvissuti e non sopravvissuti
conteggio = df['Survived'].value_counts()
print("\nConteggio di sopravvissuti e non sopravvissuti:")
conteggio
```

```python
# 8) Visualizza i nomi dei sopravvissuti
nomi_sopravvissuti = df[df['Survived'] == 1]['Name']
print("\nNomi dei sopravvissuti:")
nomi_sopravvissuti
```

```python
# 9) Calcola statistiche descrittive per le colonne numeriche del dataset
statistiche_descrittive = df.describe()
print("\nStatistiche descrittive per colonne numeriche:")
statistiche_descrittive
```

```python
# 10) Sopravvivenza per genere: Calcola la percentuale di sopravvivenza per ciascun genere
# Conteggio dei sopravvissuti per genere
sopravvissuti_per_genere = df.groupby('Sex')['Survived'].sum()
sopravvissuti_per_genere
```

```python
# Conteggio totale per genere
totale_per_genere = df['Sex'].value_counts()
totale_per_genere
```

```python
# grafico
labels=totale_per_genere.index
labels
```

```python
import matplotlib.pyplot as plt

# Grafico a torta
plt.figure(figsize=(6, 6))
plt.pie(totale_per_genere, labels=totale_per_genere.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuzione dei passeggeri per sesso')
plt.axis('equal')
plt.show()
```

```python
# Calcola la percentuale di sopravvivenza per genere
percentuale_sopravvivenza_per_genere = (sopravvissuti_per_genere / totale_per_genere) * 100
percentuale_sopravvivenza_per_genere
```

```python
# 10) Oppure --- Calcola la percentuale di sopravvivenza per genere
percentuale_sopravvivenza_per_genere = df.groupby('Sex')['Survived'].mean() * 100
percentuale_sopravvivenza_per_genere
```

```python
print(percentuale_sopravvivenza_per_genere)
# Grafico a torta
plt.figure(figsize=(6, 6))
plt.pie(percentuale_sopravvivenza_per_genere, labels=percentuale_sopravvivenza_per_genere.index, autopct='%1.2f%%', startangle=140)
plt.title('Percentuale di sopravvivenza per genere')
plt.legend()
plt.axis('equal')  # Per rendere il grafico a torta circolare
plt.show()
```

```python

# grafico a barre!!!

plt.figure(figsize=(8, 6))
plt.bar(percentuale_sopravvivenza_per_genere.index, percentuale_sopravvivenza_per_genere, color=['blue', 'green'])
plt.xlabel('Genere')
plt.ylabel('Percentuale di sopravvivenza')
plt.title('Percentuale di sopravvivenza per genere')
plt.ylim(0, 100)  # Assicura che l'asse y vada da 0 a 100%
plt.show()

```
