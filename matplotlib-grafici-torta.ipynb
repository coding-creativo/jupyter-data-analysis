# Matplotlib
Matplotlib è una libreria di visualizzazione di dati per il linguaggio di programmazione Python. È molto potente e offre una vasta gamma di grafici e plot per la creazione di visualizzazioni di dati di alta qualità. È utilizzato ampiamente in ambiti come la scienza dei dati, l'analisi statistica, l'ingegneria e molti altri settori che richiedono la rappresentazione visuale dei dati.

Matplotlib può creare tanti tipi di grafici, tra cui:

- ***Grafici a linee***: per rappresentare relazioni tra dati continui.
- ***Scatter plot***: per visualizzare punti di dati senza connessioni lineari.
- ***Grafici a torta***: per esprimere proporzioni o percentuali.
- ***Grafici a barre***: per confrontare quantità tra diverse categorie.
- ***Istogrammi***: per mostrare la distribuzione dei dati.


## PIE - GRAFICI A TORTA

Per creare un grafico a torta in Matplotlib, possiamo utilizzare la funzione ***plt.pie()***. 

Passi principali per creare un grafico scatter:

* Prepariamo i dati da visualizzare come sezioni del grafico a torta. Generalmente, è una lista di valori che rappresentano le proporzioni di ciascuna sezione.

* Utilizziamo plt.pie() passando i dati come argomenti principali. È possibile specificare opzioni aggiuntive come etichette, colori, percentuali, esplosione delle sezioni, ecc.

* Aggiungere etichette alle sezioni, un titolo al grafico, e una legenda se necessario.

* Utilizziamo plt.show() per visualizzare il grafico.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## GRAFICO A TORTA


```python
# Dati per il grafico a torta
labels = np.array(['A', 'B', 'C', 'D'])
sizes = np.array([15, 30, 45, 10]) # Percentuali

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

explode = (0.1, 0.2, 0, 0)  # Esplosione del primo cuneo - del 10% - gli altri rimangono a 0

plt.figure(figsize=(8, 6))

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
# autopct formatta il testo delle etichette che mostrano le percentuali relative a ciascuna fetta nel grafico a torta.
    # %1.1f%% indica il formato della stringa di formattazione:
    #%: Indica che il valore deve essere formattato come percentuale.
    #1.1f: Specifica che il numero deve essere formattato con una cifra decimale.
    #Quindi, %1.1f%% significa che i numeri devono essere formattati con una cifra decimale e seguiti dal simbolo percentuale %.

plt.axis('equal')  # Equalizza l'asse per renderlo un cerchio piuttosto che un'ellisse

plt.title('Esempio di grafico a torta')
plt.show()
```

```python
# Dati di esempio
data = {'Categoria': ['Elettronica', 'Abbigliamento', 'Alimentari', 'Libri', 'Sport'],
        'Vendite (%)': [25, 20, 30, 15, 10]}

sales_data = pd.DataFrame(data)

# Estraiamo i dati per il grafico a torta
labels = sales_data['Categoria']
sizes = sales_data['Vendite (%)']

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']
explode = (0.1, 0.1, 0.1, 0.1, 0.2)  # Esplosione del primo cuneo

# Crea il grafico a torta
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equalizza l'asse per renderlo un cerchio

plt.title('Distribuzione percentuale delle vendite per categoria')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Chrome', 'Firefox', 'Safari', 'Edge', 'Altro']
sizes = [65, 10, 10, 7, 8]  # Percentuali di utilizzo dei browser

plt.figure(figsize=(8, 8))

# Crea il grafico a torta con esplosione della prima fetta (Chrome)
patches, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=(0.1, 0, 0, 0, 0), shadow=True)

# Aggiungi una legenda con le etichette dei browser
plt.legend(patches, labels, loc="best", fontsize='medium')

plt.title('Distribuzione del Mercato dei Browser')
plt.show()
```
