# Matplotlib
Matplotlib è una libreria di visualizzazione di dati per il linguaggio di programmazione Python. È molto potente e offre una vasta gamma di grafici e plot per la creazione di visualizzazioni di dati di alta qualità. È utilizzato ampiamente in ambiti come la scienza dei dati, l'analisi statistica, l'ingegneria e molti altri settori che richiedono la rappresentazione visuale dei dati.

Matplotlib può creare tanti tipi di grafici, tra cui:

- ***Grafici a linee***: per rappresentare relazioni tra dati continui.
- ***Scatter plot***: per visualizzare punti di dati senza connessioni lineari.
- ***Grafici a torta***: per esprimere proporzioni o percentuali.
- ***Grafici a barre***: per confrontare quantità tra diverse categorie.
- ***Istogrammi***: per mostrare la distribuzione dei dati.


## SCATTER PLOT

è particolarmente utile quando si desidera visualizzare la relazione tra due variabili continue.
* Relazioni tra parametri vitali dei pazienti (pressione sanguigna, battito cardiaco, etc.): Utilizzato in ambito medico per trovare correlazioni tra vari parametri.
* Scienze Ambientali e Geografiche: Per visualizzare la distribuzione spaziale di fenomeni naturali come la temperatura, la precipitazione, l'inquinamento, ecc.
* Economia e Finanza: Per esaminare le correlazioni tra variabili economiche come il prezzo delle azioni e il volume di scambio, la relazione tra indicatori macroeconomici come PIL, tasso di disoccupazione, ecc.
* Marketing e Ricerca di Mercato: Per esaminare le relazioni tra variabili di mercato come il prezzo e la domanda di un prodotto, il livello di soddisfazione del cliente e la frequenza degli acquisti, ecc.

Per creare un grafico scatter in Matplotlib, possiamo utilizzare la funzione ***plt.scatter()***. 

Passi principali per creare un grafico scatter:

* Prepariamo i dati da visualizzare come punti scatter. Generalmente, sono due array o liste, uno per l'asse X e uno per l'asse Y, che rappresentano le coordinate dei punti nel grafico.

* Utilizziamo plt.scatter() passando i dati sull'asse X e sull'asse Y come argomenti principali. È possibile specificare opzioni aggiuntive come il colore dei punti, le dimensioni dei punti, i colori delle facce, i marcatori e altro ancora.

* Aggiungiamo etichette agli assi, un titolo al grafico, una legenda se necessario e abilitiamo la griglia per migliorare la leggibilità del grafico.

* Utilizziamo plt.show() per visualizzare il grafico.

https://matplotlib.org/stable/gallery/color/named_colors.html

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Scatter plot

```python
# Dati di esempio: Prezzo del prodotto e quantità domandata
prezzo = np.array([10, 15, 20, 25, 30])  # Prezzi in euro
quantità =  np.array([100, 80, 60, 40, 20])  # Quantità domandata

# Creazione dello scatter plot
plt.figure(figsize=(8, 6))

plt.scatter(prezzo, quantità, color='skyblue', label='Dati di mercato')

plt.title('Relazione tra Prezzo e Domanda di un Prodotto')

plt.xlabel('Prezzo (Euro)')
plt.ylabel('Quantità Domandata')

plt.grid(True)

plt.legend()
plt.show()
```

## Definire una mappa di colori

```python
# Definizione dei colori per ogni punto
colors = ['blue', 'green', 'red', 'purple', 'orange']

# Creazione dello scatter plot
plt.figure(figsize=(8, 6))

plt.scatter(prezzo, quantità, color=colors, label='Dati di mercato')

plt.title('Relazione tra Prezzo e Domanda di un Prodotto')

plt.xlabel('Prezzo (Euro)')
plt.ylabel('Quantità Domandata')
plt.grid(True)

plt.tight_layout() #adatta il grafico automaticamente

plt.legend()
plt.show()
```

```python
np.random.seed(0)  

altezze = np.random.normal(loc=167, scale=9.52, size=100)  
pesi = altezze * 0.45 + np.random.normal(loc=0, scale=2.27, size=100)
''' generiamo:
    --> altezze in centimetri, con una media di 167 cm e una deviazione standard di 9.52 cm
    --> generiamo pesi in chilogrammi basati su una relazione approssimativa con l'altezza, aggiungendo un po' di variabilità casuale.
    Il parametro size=100 nella funzione np.random.normal indica che vogliamo generare 100 valori casuali. 
    Questi valori rappresentano le 100 osservazioni delle altezze e dei pesi nel dataset.
''' 
# Creiamo un dataframe di esempio
data = {
    'Altezza (cm)': altezze,
    'Peso (kg)': pesi
}

df = pd.DataFrame(data)

# Visualizziamo le prime righe del dataframe
df.head()
```

```python
# Creazione dello scatter plot per la relazione tra altezza e peso
plt.figure(figsize=(10, 6))

colori = np.random.rand(100)  # Genera 100 colori casuali tra 0 e 1

# Scatter plot con altezze in cm e pesi in kg
# https://matplotlib.org/stable/gallery/color/colormap_reference.html
scatter = plt.scatter(df['Altezza (cm)'], df['Peso (kg)'], c=colori, cmap='inferno', alpha=0.5)

plt.title('Scatter plot: Altezza vs Peso')

plt.xlabel('Altezza (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)

''' Alcune Colormap Disponibili: 
        'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds'
        'YlOrBr', 'YlOrRd','OrRd','PuRd','RdPu','BuPu','GnBu','PuBu','YlGnBu','PuBuGn','BuGn','YlGn'
'''

# Aggiunta di una barra dei colori
cbar = plt.colorbar(scatter)

# ottimizziamo automaticamente il layout del grafico per evitare sovrapposizioni tra elementi come titoli, etichette degli assi, 
#legende e altri componenti del grafico.
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dati casuali per le tre variabili
np.random.seed(0)
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot tridimensionale
ax.scatter(x, y, z, c=x, cmap='rainbow', marker='o')

# Etichette degli assi
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Titolo del grafico
ax.set_title('Grafico Scatter Tridimensionale')

plt.tight_layout()
plt.show()
```
